```python
import cmdstanpy
import numpy as np
import pandas as pd
```

```python
column_names = dict(
    Subject = 'ids',
    Time = 'time',
    Concentration_scaled = 'y',
    Sex = 'sex',
    Age = 'age',
    Creatinine = 'creatinine',
    study='study',
    Weight='weight'
)

rommel_df = pd.read_csv('data/rommels_apixiban_data.csv').assign(D = 5, study = 'rommel').rename(columns = column_names)
rommel_df.head()
```

```python
column_names = {
    "New Subject ID" : 'ids',
    'hrs_post_dose' : 'time',
    'concentration_ng_per_ml':'y',
    'sex' : 'sex',
    'enrollment_age' : 'age',
    'creatinine_umol_per_L' : 'creatinine',
    'dose_mg' : 'D',
    'study':'study',
    'weight_kg':'weight'
}

new_df = pd.read_csv('data/apixaban_2020_data.csv').assign(study = 'new').rename(columns = column_names)
new_df.head()
```

```python
df = pd.concat((new_df, rommel_df)).loc[:, column_names.values() ]

ids = dict()

for i,j in enumerate(df.ids.unique()):
    ids[j]=i
    
    
df['new_ids'] = df.ids.apply(lambda x: ids[x])
```

```python
model_data = dict(
    N = df.shape[0],
    n_subjects = df.new_ids.nunique(),
    subjectids = df.new_ids.tolist(),
    yobs = df.y.tolist(),
    times = (1000*df.time).tolist(),
    D = df.D.values.tolist(),
    ppc_t = np.linspace(0,12,8).tolist()

)
```

```python
model_data

import json

with open('data.json', 'w') as fp:
    json.dump(model_data, fp)
```
