```python
import cmdstanpy
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = pd.read_csv('data/combined_data.csv')
data['subjectids'] = pd.Categorical(data.subjectids).codes + 1
data['sex'] = data.sex.apply(lambda x: 1 if x == 'male' else 0)
data['yobs'] = data.yobs/1000

old_study = data.query('study=="old"')
new_study = data.query('study=="new"')

scaler = StandardScaler()

scaler.fit(old_study.loc[:,['age','weight','creatinine']].values)

data.loc[:,['age','weight','creatinine']] = scaler.transform(data.loc[:,['age','weight','creatinine']].values)

subjects = data.drop_duplicates(subset = 'subjectids')

model_data = dict(
    N = data.shape[0],
    subjectids = data.subjectids.tolist(),
    n_subjects = data.subjectids.nunique(),
    time = data.time.tolist(),
    sex = subjects.sex.tolist(),
    weight = subjects.weight.tolist(),
    creatinine = subjects.creatinine.tolist(),
    D = subjects.D.tolist(),
    yobs = (data.yobs).tolist()
    
)
```

```python
model = cmdstanpy.CmdStanModel(stan_file = 'model_dev_2.stan')
```

```python
fit = model.sample(data = model_data, chains=4)
```

```python
pred = fit.get_drawset('C').mean().to_frame('pred').reset_index()

f = pd.concat((pred, data),axis=1)


g = sns.FacetGrid(data = f, col='study', height = 5)

x = np.linspace(0,1)
g.map_dataframe(plt.scatter, 'yobs','pred')
```
